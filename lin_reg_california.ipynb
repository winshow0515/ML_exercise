```python
import torch

class LinearRegression(torch.nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        self.linear = torch.nn.Linear(input_size, 512)
        self.output = torch.nn.Linear(512, output_size)
    def forward(self, x):
        out = self.output(self.linear(x))
        return out


# linear = torch.nn.Linear(6, 1)
# print(linear.weight)
# print(linear.bias)
# x = torch.Tensor([1, 0, 0, 0, 0, 0])
# print(linear(x))
```

```python
criterion = torch.nn.MSELoss()
print(criterion(torch.tensor([1., 0]), torch.tensor([9., 1])))
# torch.mean((y_pred - y)**2)
```

```python
# Deploy linear algorithm by torch

input_size = 6
output_size = 1
model = LinearRegression(input_size, output_size)

criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-9)

for epoch in range(10):
    y_pred = model(X)
    loss = criterion(y_pred, y)
    print(epoch, loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
