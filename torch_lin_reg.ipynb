```python
import numpy as np
import matplotlib.pyplot as plt
# BMI = 17~23
BMI = np.random.randn(100) * 4 + 22
H = np.random.randn(100) * 8 + 170
W = BMI * (H / 100) ** 2

def plot(w, b):
    plot_x = np.array([140, 195])
    plt.plot(plot_x, plot_x * w + b, c='red')
    plt.scatter(H, W)
    plt.xlim(140, 195)
    plt.ylim(30, 120)
    plt.xlabel('Height')
    plt.ylabel('Weight')
    plt.title("Random Weight and Height")
    plt.show()

plot(1.2, -140)
```

```python
def get_gradient(w, b):
    w_gradient_2 = 2 * np.mean((H * w + b - W) * H)
    b_gradient_2 = 2 * np.mean(H * w + b - W)
    return w_gradient_2, b_gradient_2
print("w_gradient, %f b_gradient = %f" % get_gradient(0, 0))
```

```python
import torch
H, W = torch.Tensor(H), torch.Tensor(W)
w = torch.Tensor([0]); w.requires_grad = True
b = torch.Tensor([0]); b.requires_grad = True
```

```python

# Calculate Gradient by torch
pred = H * w + b
loss = torch.mean((W - pred) ** 2)

w_gradient, b_gradient = torch.autograd.grad(loss, [w, b])
print(w_gradient, b_gradient)
print(loss.item())

plot(w.item(), b.item())
pred = H * w + b
w = w - 0.000001 * w_gradient
b = b - 0.000001 * b_gradient
# print(pred)
```
