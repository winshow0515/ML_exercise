```python
import numpy as np

# BMI = 17~23
BMI = np.random.randn(100) * 4 + 22
H = np.random.randn(100) * 8 + 170
W = BMI * (H / 100) ** 2

import matplotlib.pyplot as plt
# import matplotlib as mpl
# mpl.rcParams['axes.unicode_minus'] = False
plt.scatter(H, W)
plt.xlim(140, 195)
plt.ylim(30, 120)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title("Random Weight and Height")

w1, b1 = 1.2, -140
w2, b2 = 1.3, -160

pred_1 = H * w1 + b1
loss_1 = np.mean((W - pred_1) ** 2)
pred_2 = H * w2 + b2
loss_2 = np.mean((W - pred_2) ** 2)
print("Loss 1:", loss_1)
print("Loss 2:", loss_2)

plot_x = np.array([140, 195])
plt.plot(plot_x, plot_x * w1 + b1, c='red')
plt.plot(plot_x, plot_x * w2 + b2, c='green')
plt.show()
```

```python
def get_loss(w, b):
    pred_W = H ** 2 * w + b
    return np.mean((W-pred_W)**2)

def get_gradient(w, b):
    w_gradient_2 = 2 * np.mean((H ** 2 * w + b - W) * H ** 2)
    b_gradient_2 = 2 * np.mean(H ** 2 * w + b - W)
    return w_gradient_2, b_gradient_2
print(get_gradient(0, 0))
def plot(w, b):
    plot_x = np.array([140, 195])
    plt.plot(plot_x, plot_x ** 2 * w + b, c='red')
    plt.scatter(H, W)
    plt.xlim(140, 195)
    plt.ylim(30, 120)
    plt.xlabel('Height')
    plt.ylabel('Weight')
    plt.title("Random Weight and Height")
    plt.show()

w = 0
b = 0
plot(1.2, -140)
```

```python
w_gradient, b_gradient = get_gradient(w, b)
learning_rate = 0.0000000001
w = w - learning_rate * w_gradient
b = b - learning_rate * b_gradient

print(f"w = {w}, b = {b}, Loss = {get_loss(w, b)}")
plot(w, b)
# 115
```
